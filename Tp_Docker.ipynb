1. Récupération du projet

    Pour récupérer le projet depuis Github :

```python
# Cloner le repository
git clone https://github.com/sofian-ras/tp-Docker.git

# Aller dans le dossier du projet
cd tp-Docker
```

2. Lancement du projet sous Docker

Pour construire et exécuter les conteneurs read et write :

```python
# Construire et lancer les conteneurs
docker-compose up --build
```

Pour exécuter les scripts Python directement dans les conteneurs :

```python
# Entrer dans le conteneur write
docker exec -it write_container bash
python manip_data.py

# Entrer dans le conteneur read
docker exec -it read_container bash
python read_data.py
```

Les conteneurs partagent le dossier /app/data grâce au volume data_volume, donc toutes les modifications dans write sont visibles dans read.

3. Utilisation des applications
3.1 Write (manip_data.py)

Fonctionnalités :

**Ajouter un film**

- Modifier un film (en utilisant son ID)

- Supprimer un film (en utilisant son ID)

**Entrées valides :**

- Titre : texte (ex : Titanic)

- Année : entier (ex : 1997)

- Genre : texte (ex : Drame, Action)

- Âge minimum : entier (ex : 12)

Exemple d’exécution :

```python
python manip_data.py
Titre du film : Titanic
Année de production : 1997
Genre : Drame
Âge minimum : 12
```

Le film est ajouté dans le CSV et peut être consulté depuis read.

3.2 Read (read_data.py)

**Fonctionnalités :**

- Lister tous les films

- Filtrer par genre

- Filtrer par année

- Rechercher par titre

- Filtrer par limite d’âge

- Filtrer par période (entre deux années)

Exemple d’exécution :

```python
python read_data.py
~~~~ filtrer les films ~~~~
Filtrer par (g)enre ou (a)nnée ? g
Entrez le genre : Drame
ID: 31 | Titre: Titanic | Année: 1997 | Âge min: 12
```

Si aucun film ne correspond, un message vous prévient.

3.3 Notes importantes

Les fichiers CSV sont stockés dans /app/data dans chaque conteneur.

Grâce au volume data_volume, read et write voient le même CSV.

Toujours vérifier les entrées : année et âge doivent être des nombres, genre et titre du texte.

Les IDs des films sont automatiques et incrémentés à chaque ajout.
